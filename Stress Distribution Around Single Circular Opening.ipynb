**Importing Necessary Libraries**

```
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

**Defining Kirsch Relations**

```
#Radial Stress/Po
def sig_rr(ratio, k, theta):
  sigrr = (1/2)*(1+k)*(1-(ratio**2)) - (1/2)*(1-k)*(1 + 3*(ratio**4) - 4*(ratio**2))*math.cos(2*theta)
  return(sigrr)

#Tangential Stress/Po
def sig_tt(ratio, k, theta):
  sigtt = (1/2)*(1+k)*(1+(ratio**2)) + (1/2)*(1-k)*(1 + 3*((ratio**4)))*math.cos(2*theta)
  return(sigtt)

#Shear Stress/Po
def sig_rt(ratio, k, theta):
  sigrt = (1/2)*(1-k)*(1-3*(ratio**4) + 2*(ratio**2))*math.sin(2*theta)
  return(sigrt)
```

**Defining Range of r/a & k**

Range of *r/a* Values taken from **1** to **5** with increment value of **0.001**. Similarly values of *k* was taken from **0** to **1** with increment value of **0.3**

```
#Defining Range of r/a
ratio_inverse = np.arange(1,5,0.001)
#Defining Range of k
k = np.arange(0,1,0.3)
```

Define Theta

```
theta_case1 = math.radians(0)
```

**SAMPLE DATA POINTS**

```
ratio_inverse_sample = np.arange(1,5,1)
sig_rr_values_sample = [[0 for j in range(k.size)] for i in range(ratio_inverse_sample.size)]
sig_tt_values_sample = [[0 for j in range(k.size)] for i in range(ratio_inverse_sample.size)]

print("--------------PRINTING SAMPLE DATA VALUES----------")
print("\n")
for j in range(k.size):
   for i in range(ratio_inverse_sample.size):
     a_by_r = 1/ratio_inverse_sample[i]
     sig_tt_values_sample[i][j] = sig_tt(a_by_r, k[j], theta_case1)
     sig_rr_values_sample[i][j] = sig_rr(a_by_r,k[j], theta_case1)
     print("r/a : " + str(ratio_inverse[i]) + "  k-value : " + str(k[j]) + "  Tangential Stress/Po : " + str(sig_tt_values_sample[i][j]))
```

**Define Data Points for Plotting**

```
sig_rr_values = [[0 for j in range(k.size)] for i in range(ratio_inverse.size)]
sig_tt_values = [[0 for j in range(k.size)] for i in range(ratio_inverse.size)]

for j in range(k.size):
   for i in range(ratio_inverse.size):
     a_by_r = 1/ratio_inverse[i]
     sig_tt_values[i][j] = sig_tt(a_by_r, k[j], theta_case1)
     sig_rr_values[i][j] = sig_rr(a_by_r,k[j], theta_case1)
```

**Stress Distribution for Single Circular Opening for theta=0**


```
from matplotlib.pyplot import figure

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,15))
for i, ax in enumerate(axes.flat, start=1):
    ax.set_title('Stress Distribution for Theta = 0 and k = {}'.format(k[i-1]))
    sig_tt_vals =  [row[i-1] for row in sig_tt_values]
    sig_rr_vals =  [row[i-1] for row in sig_rr_values]
    ax.plot(ratio_inverse,sig_tt_vals)
    ax.plot(ratio_inverse, sig_rr_vals)
    ax.set_xlabel('r/a value')
    ax.set_ylabel('Tangential Stress/Po || Radial Stress/Po')
    ax.legend(['Tangential Sress/Po','Radial Stress/Po'])

fig.tight_layout()
plt.show()
```
